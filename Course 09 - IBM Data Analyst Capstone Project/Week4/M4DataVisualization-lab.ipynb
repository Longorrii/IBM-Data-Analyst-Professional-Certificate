# Hands-on Lab


## Visualizing distribution of data


```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""
ConvertedComp = pd.read_sql_query(QUERY,conn)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(ConvertedComp['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Age
FROM master
"""
Age = pd.read_sql_query(QUERY,conn)

# Plot box plot
Age.plot(kind='box', figsize=(8, 10))
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Fetch data
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
data = pd.read_sql_query(QUERY, conn)

# Extract Age and WorkWeekHrs columns
Age = data['Age']
WorkWeekHrs = data['WorkWeekHrs']

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(Age, WorkWeekHrs, alpha=0.5)
plt.title('Scatter plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

```python
data.corr()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

# Fetch data
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
data = pd.read_sql_query(QUERY, conn)

# Extract columns
WorkWeekHrs = data['WorkWeekHrs']
CodeRevHrs = data['CodeRevHrs']
Age = data['Age']

# Plot bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(WorkWeekHrs, CodeRevHrs, s=Age, alpha=0.5)
plt.title('Bubble plot of WorkWeekHrs and CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# Fetch data
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) AS Count
FROM DatabaseDesireNextYear
WHERE DatabaseDesireNextYear IS NOT NULL
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
data = pd.read_sql_query(QUERY, conn)

# Extract database names and counts
database_names = data['DatabaseDesireNextYear']
counts = data['Count']

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=database_names, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Fetch data
QUERY = """
SELECT LanguageDesireNextYear, COUNT(*) AS Count
FROM LanguageDesireNextYear
WHERE LanguageDesireNextYear IS NOT NULL
GROUP BY LanguageDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
data = pd.read_sql_query(QUERY, conn)

# Extract database names and counts
Language_names = data['LanguageDesireNextYear']
counts = data['Count']

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=Language_names, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Fetch data
QUERY = """
SELECT COUNT(*) AS NumRespondents
FROM LanguageWorkedWith
WHERE LanguageWorkedWith LIKE '%SQL%'
"""
data = conn.execute(QUERY).fetchone()

num_respondents_sql = data[0]

print("Number of respondents who indicated that they currently work with 'SQL':", num_respondents_sql)
```

```python
# Fetch data
QUERY = """
SELECT COUNT(*) AS NumRespondents
FROM DatabaseWorkedWith
WHERE DatabaseWorkedWith = 'MySQL'
  AND NOT EXISTS (SELECT 1
                  FROM DatabaseWorkedWith AS d2
                  WHERE d2.Respondent = DatabaseWorkedWith.Respondent
                  AND d2.DatabaseWorkedWith <> 'MySQL')
"""
data = conn.execute(QUERY).fetchone()

num_respondents_mysql_only = data[0]

print("Number of respondents who indicated that they work exclusively on 'MySQL':", num_respondents_mysql_only)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Avg(WorkWeekHrs), Avg(CodeRevHrs), Age 
FROM master 
where Age between 30 and 35
group by Age
"""

z = pd.read_sql_query(QUERY, conn)
WorkWeekHrs= z['Avg(WorkWeekHrs)']
CodeRevHrs= z['Avg(CodeRevHrs)']
Age= z['Age']

fig, ax = plt.subplots()

ax.bar(Age, WorkWeekHrs, label='WorkWeekHrs')
ax.bar(Age, CodeRevHrs, bottom=WorkWeekHrs,
       label='CodeRevHrs')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# Fetch data
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age >= 45 AND Age <= 60
"""
data = pd.read_sql_query(QUERY, conn)

# Calculate median ConvertedComp for each age
median_data = data.groupby('Age').median()

# Extract median values for ConvertedComp
median_converted_comp = median_data['ConvertedComp']

# Define age groups
age_groups = median_data.index

# Plot median ConvertedComp
plt.figure(figsize=(10, 6))
plt.plot(age_groups, median_converted_comp, marker='o', color='blue', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
# Fetch data
QUERY = """
SELECT MainBranch, COUNT(*) AS Count
FROM master
WHERE MainBranch IS NOT NULL
GROUP BY MainBranch
"""
data = pd.read_sql_query(QUERY, conn)

# Sort data by count in descending order
data.sort_values(by='Count', ascending=False, inplace=True)

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(data['MainBranch'], data['Count'], color='skyblue')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branches')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest count at the top
plt.grid(axis='x')  # Add grid lines only along the x-axis
plt.show()
```

```python
# Fetch data
QUERY = """
SELECT DevType, COUNT(*) AS RespondentCount
FROM DevType
WHERE DevType IS NOT NULL
GROUP BY DevType
ORDER BY RespondentCount DESC
"""
data = pd.read_sql_query(QUERY, conn)

# Assuming the first row with the highest RespondentCount represents the majority
majority_dev_type = data['DevType'].iloc[0]
majority_count = data['RespondentCount'].iloc[0]

print("Majority of the survey responders are", majority_dev_type, "with a count of", majority_count)
```
